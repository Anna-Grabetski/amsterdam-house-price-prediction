# üè° –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è —Ü–µ–Ω –Ω–∞ –∂–∏–ª—å—ë –≤ –ê–º—Å—Ç–µ—Ä–¥–∞–º–µ

–≠—Ç–æ—Ç –Ω–æ—É—Ç–±—É–∫:
- –ó–∞–≥—Ä—É–∂–∞–µ—Ç **–æ—á–∏—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ**
- –ü—Ä–∏–º–µ–Ω—è–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
- –°—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç –∏—Ö –ø–æ –º–µ—Ç—Ä–∏–∫–∞–º
- –°–æ—Ö—Ä–∞–Ω—è–µ—Ç –æ–±—É—á–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª–∏


üîπ –ò–º–ø–æ—Ä—Ç—ã –∏ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from src.preprocessing import load_cleaned_data
from src.train import scale_features, train_test_split_data
from src.evaluate import evaluate_model
from src.utils import save_model_named

sns.set(style="whitegrid")
```

üîπ –ó–∞–≥—Ä—É–∑–∫–∞ –æ—á–∏—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö

```python
df = load_cleaned_data(path="data/cleaned.csv")
df.head(2)
```

üîπ –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é

```python
FEATURES = ["area", "rooms", "price_per_sqm", "area_per_room", "lat_bin", "lon_bin", "distance_to_center_km", "zip_prefix"]
TARGET = "price"

X = df[FEATURES]
y = df[TARGET]
```

üîπ –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ —Å–ø–ª–∏—Ç

```python
X_scaled = scale_features(X.copy(), FEATURES)
X_train, X_test, y_train, y_test = train_test_split_data(X_scaled, y)
```

üîπ –û–±—ä—è–≤–ª–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

```python
models = {
    "LinearRegression": LinearRegression(),
    "Ridge(alpha=1.0)": Ridge(alpha=1.0),
    "DecisionTree(depth=5)": DecisionTreeRegressor(max_depth=5),
    "RandomForest(n=50)": RandomForestRegressor(n_estimators=50, random_state=42)
}
```

üîπ –û–±—É—á–µ–Ω–∏–µ, –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ, –æ—Ü–µ–Ω–∫–∞ –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ

```python
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    metrics = evaluate_model(y_test, y_pred)
    metrics["Model"] = name
    results.append(metrics)

    safe_name = name.lower().replace("(", "_").replace(")", "").replace("=", "_").replace(" ", "").replace(",", "")
    save_model_named(model, safe_name)
```

üîπ –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

```python
df_results = pd.DataFrame(results).set_index("Model")
df_results.style.background_gradient(cmap="Blues").format("{:.2f}")
```

üîπ –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: —Ä–µ–∞–ª—å–Ω–æ–µ vs –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–µ –∏ –≤–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # üßÆ –ü–æ–¥—Å—á—ë—Ç –≤–∞–∂–Ω–æ—Å—Ç–µ–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
    elif hasattr(model, 'coef_'):
        importances = np.abs(model.coef_)
    else:
        importances = None

    # üìä –î–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä—è–¥–æ–º
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # –ì—Ä–∞—Ñ–∏–∫ 1: –†–µ–∞–ª—å–Ω—ã–µ vs –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    sns.scatterplot(ax=axes[0], x=y_test, y=y_pred, alpha=0.7)
    axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    axes[0].set_title(f"{name} | –†–µ–∞–ª—å–Ω–æ–µ vs –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–µ")
    axes[0].set_xlabel("–†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è")
    axes[0].set_ylabel("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è")

    # –ì—Ä–∞—Ñ–∏–∫ 2: –í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    if importances is not None:
        sns.barplot(ax=axes[1], x=importances, y=X_train.columns, orient='h')
        axes[1].set_title(f"{name} | –í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
        axes[1].set_xlabel("Importance")
    else:
        axes[1].set_visible(False)

    plt.tight_layout()
    plt.show()
```
